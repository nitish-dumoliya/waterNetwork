```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_network(G, edges, title, ax, demands):
    pos = {1: (3, 2), 2: (2, 2), 3: (1, 2), 4: (2, 1), 5: (1, 0)}
    
    # Node colors
    node_colors = ["red" if node == 1 else 'lightblue' for node in G.nodes()]
    
    # Draw graph
    nx.draw(
        G, pos, with_labels=True, node_color=node_colors, edge_color='black', width=2,
        node_size=1200, font_size=16, font_weight='bold',
        ax=ax, arrows=True, arrowsize=25, arrowstyle='->',
        edgecolors='black', linewidths=1.5
    )

    # Edge labels
    edge_labels = {(u, v): f"{d}" for u, v, d in edges}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='blue', font_size=13, ax=ax)

    # Node demands with arrow and label below it
    for node, demand in demands.items():
        # if node == 1:
        #     continue  # Skip the source node

        x, y = pos[node]
        arrow_x, arrow_y = x - 0.08, y - 0.2

        ax.text(
            arrow_x, arrow_y,
            r'$\Leftarrow$',
            fontsize=16,
            color='gray',
            rotation=45,
            ha='center', va='center'
        )

        ax.text(
            arrow_x - 0.08, arrow_y - 0.02,
            f"{demand}",
            fontsize=11,
            color='black',
            ha='center', va='top'
        )

    # Source label
    ax.text(pos[1][0], pos[1][1] + 0.25, "Source", fontsize=14, fontweight='bold', ha='center', color='blue')

    ax.set_title(title, fontsize=16, fontweight='bold', color='purple')
    ax.axis('off')

# Node demands
demands = {1: "-40", 2: "10", 3: "10", 4: "10", 5: "10"}

# ---------- Figure 1: Branch Network ----------
fig1, ax1 = plt.subplots(figsize=(6, 5))
edges_branch = [(1, 2, 40), (2, 3, 10), (2, 4, 20), (4, 5, 10)]
G_branch = nx.DiGraph()
G_branch.add_weighted_edges_from(edges_branch)
draw_network(G_branch, edges_branch, "Branch Network", ax1, demands)
plt.tight_layout()
plt.show()

# ---------- Figure 2: Loop Networks (2 subplots) ----------
fig2, axes = plt.subplots(2, 1, figsize=(6, 10))

# Loop Network 1
edges_loop1 = [(1, 2, 40), (2, 3, 15), (2, 4, 15), (4, 5, 5), (3, 5, 5)]
G_loop1 = nx.DiGraph()
G_loop1.add_weighted_edges_from(edges_loop1)
draw_network(G_loop1, edges_loop1, "Loop Network 1", axes[0], demands)

# Loop Network 2
edges_loop2 = [(1, 2, 40), (2, 3, 25), (2, 4, 5), (5, 4, 5), (3, 5, 15)]
G_loop2 = nx.DiGraph()
G_loop2.add_weighted_edges_from(edges_loop2)
draw_network(G_loop2, edges_loop2, "Loop Network 2", axes[1], demands)

plt.tight_layout()
fig2.savefig("loopedNetwork.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create the graph
G = nx.DiGraph()

# Add nodes: (label, type)
nodes = {
    'R1': 'reservoir',
    'J1': 'junction',
    'J2': 'junction',
    'J3': 'junction',
    'V1': 'valve',
    'F1': 'fire_hydrant',
    'S1': 'service_connection'
}

# Add nodes to graph
for node, ntype in nodes.items():
    G.add_node(node, type=ntype)

# Add edges (pipes)
edges = [
    ('R1', 'J1'),
    ('J1', 'J2'),
    ('J2', 'J3'),
    ('J2', 'V1'),
    ('J1', 'F1'),
    ('J3', 'S1')
]
G.add_edges_from(edges)

# Position layout
pos = {
    'R1': (0, 2),
    'J1': (1, 2),
    'J2': (2, 2),
    'J3': (3, 2),
    'V1': (2, 1),
    'F1': (1, 3),
    'S1': (3, 1)
}

# Define styles
node_colors = {
    'reservoir': 'blue',
    'junction': 'gray',
    'valve': 'red',
    'fire_hydrant': 'orange',
    'service_connection': 'green'
}
node_shapes = {
    'reservoir': 's',           # square
    'junction': 'o',            # circle
    'valve': '^',               # triangle
    'fire_hydrant': 'D',        # diamond
    'service_connection': 'v'   # inverted triangle
}

# Draw nodes by type
for ntype, shape in node_shapes.items():
    nodelist = [n for n in G.nodes if G.nodes[n]['type'] == ntype]
    nx.draw_networkx_nodes(G, pos,
                           nodelist=nodelist,
                           node_color=node_colors[ntype],
                           node_shape=shape,
                           label=ntype.capitalize(),
                           node_size=800)

# Draw edges (pipes)
nx.draw_networkx_edges(G, pos, width=2, edge_color='black', arrows=True)
nx.draw_networkx_labels(G, pos)

# Legend and display
plt.legend(scatterpoints=1)
plt.axis('off')
plt.title("Water Distribution Network Layout")
plt.tight_layout()
plt.show()
```
