```python
import networkx as nx
import random

def generate_spanning_trees(graph, num_trees=100):
    spanning_trees = []
    edges = list(graph.edges())

    # Generate spanning trees
    for _ in range(num_trees):
        random.shuffle(edges)  # Shuffle edges to get random spanning trees
        subgraph = nx.Graph()
        subgraph.add_edges_from(edges)
        
        # Extract a spanning tree from the shuffled subgraph
        if nx.is_connected(subgraph):
            tree = nx.minimum_spanning_tree(subgraph)
            spanning_trees.append(tree)
    
    return spanning_trees

# Example: Create a random graph and generate spanning trees
G = nx.gnp_random_graph(10, 0.5)  # Create a random graph with 10 nodes and probability 0.5
spanning_trees = generate_spanning_trees(G, num_trees=100)

# Check and print the number of spanning trees generated
print(f"Generated {len(spanning_trees)} spanning trees")
# plt.figure(figsize=(10, 8))
nx.draw_spectral(spanning_trees, with_labels=True)
plt.show()
```
